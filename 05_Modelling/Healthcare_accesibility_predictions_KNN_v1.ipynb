```python
import pandas as pd

data1 = pd.read_csv("final_healthcare_facility_dataset.csv")
data1.head()

```

```python
data2 = pd.read_csv("population-by-cities-data.csv")
data2.head()
```

```python


merged_data = pd.merge(data1, data2, left_on='City', right_on='city', how='inner')
merged_data.head()
```

```python


merged_data = merged_data.drop('city', axis=1)
merged_data.head()
```

```python


merged_data = merged_data.drop(['Unnamed: 0', 'name'], axis=1)
merged_data.head()
```

```python


merged_data = merged_data.drop('addr_city', axis=1)
merged_data.head()
```

```python
# latitude_city_pop and longitude_city_pop

merged_data = merged_data.rename(columns={
    'latitude': 'latitude_city_pop',
    'longitude': 'longitude_city_pop'
})
merged_data.head()
```

```python
#drop column name_english from merged data

merged_data = merged_data.drop('name_english', axis=1)
merged_data.head()
```

```python
# distance from the healthcare using latitude and longitude and the city population
# using latitude_city_pop and longitude_city_pop

import pandas as pd
from geopy.distance import geodesic


def compute_distance(row):
    coords_1 = (row['Latitude'], row['Longitude'])
    coords_2 = (row['latitude_city_pop'], row['longitude_city_pop'])
    return geodesic(coords_1, coords_2).km

merged_data['distance'] = merged_data.apply(compute_distance, axis=1)
merged_data.head()
```

```python
#a new feature called accesible considering if distance is equal or less than 5 km

import pandas as pd
from geopy.distance import geodesic




# he 'accessible' feature
merged_data['accessible'] = merged_data['distance'].apply(lambda x: 1 if x <= 5 else 0)

merged_data.head()
```

```python

X = merged_data_copy.drop('accessible', axis=1)
y  = merged_data_copy['accessible']

print(X.head())
```

```python
# KNN machine learning model to predict feature called accesible

import pandas as pd
from geopy.distance import geodesic
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,classification_report

   # Identify non-numeric columns
non_numeric_cols = X.select_dtypes(exclude=['number']).columns

   # Apply one-hot encoding to non-numeric features
X_encoded = pd.get_dummies(X, columns=non_numeric_cols, drop_first=True)
X_encoded = X_encoded.astype(int)
print(X_encoded.head())

print(X_encoded.dtypes)
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)  # Adjust test_size as needed

# Initialize and train the KNN model
knn_model = KNeighborsClassifier(n_neighbors=3, metric = 'manhattan', weights = 'distance' )  # Adjust n_neighbors as needed
knn_model.fit(X_train, y_train)

# predictions on the test set
y_pred = knn_model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy of KNN model: {accuracy}")

print("classification report", classification_report(y_test, y_pred))



```

```python
# hyperparrameter optimization
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV


knn = KNeighborsClassifier()

param_grid = {
    'n_neighbors': [3, 5, 7, 9],
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean','haversine' ,'manhattan', 'minkowski'] }

#Grid Search with Cross-Validation
grid_search = GridSearchCV(knn, param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_encoded, y)

# Best parameters and score
print("Best Parameters:", grid_search.best_params_)
print("Best Score:", grid_search.best_score_)
```
