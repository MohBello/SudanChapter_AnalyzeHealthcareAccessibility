```python
def flatten_sentinel_data(df, column, region_col, geom_col):
    records = []
    for x in range(len(df)):

        val = eval(df.loc[x, column])
        
        if isinstance(val, dict):
            
            for key, values in val.items():
                for item in values:
                    
                    records.append({
                        'date': pd.to_datetime(item['interval_from']),
                        'value': item['ndvi_mean'],
                        'region': df.loc[x,region_col],
                        'geometry': df.loc[x, geom_col]
                    })
    return pd.DataFrame(records)
```

```python
ndvi_data = flatten_sentinel_data(df, 'SENTINEL_2_NDVI_L3', 'NAME_L3', 'GEOMETRY_L3')
ndwi_data = flatten_sentinel_data(df, 'SENTINEL_2_NDWI_L3', 'NAME_L3', 'GEOMETRY_L3')
```

```python
ndvi_trend = ndvi_data.groupby(['date', 'region'])['value'].mean().reset_index()
ndwi_trend = ndwi_data.groupby(['date', 'region'])['value'].mean().reset_index()
```
