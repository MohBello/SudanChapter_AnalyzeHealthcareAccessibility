Analyzing Health indicators for Sudan - extracted from this [file](https://github.com/OmdenaAI/SudanChapter_AnalyzeHealthcareAccessibility/blob/main/03_Data_preprocessing/demographics/demographicsAlt.csv) :

- Years from 2000
- Indicators containing the word Health
- Dropping the non useful columns

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/drive/MyDrive/Omdena-Health_Sudan/demographic-health-indicators.csv')
```

```python
data.info()
```

```python
# plot indicators
data = data.transpose()
data.columns = data.iloc[0]
data = data.drop(data.index[0])
data.rename(columns={'Indicator': 'Year'}, inplace=True)
data.head()
```

```python
data.reset_index(inplace=True)
data.rename(columns={'index': 'Year'}, inplace=True)
data.head()
```

```python
indicators = [column for column in data.columns if column != 'Year']
indicators
```

```python
data.plot(x='Year', y=[column for column in data.columns if column != 'Year'], kind='line')
# plt.figure(figsize=(20, 30)) # size of the plot
# data.plot(x='Year', y=['Current health expenditure per capita (current US$)',
#                        'Domestic private health expenditure per capita (current US$)',
#                        'Domestic general government health expenditure per capita (current US$)'], kind='line')

plt.gcf().set_size_inches(20, 10)

plt.xlabel('Year')  # Set the x-axis label
plt.ylabel('Values')  # Set the y-axis label
plt.title('Health Indicators Over Time')  # Set the title
plt.legend(loc='upper left')  # Display the legend
# plt.grid(True)  # Add a grid for better readability
# plt.show()  # Show the plot
```
