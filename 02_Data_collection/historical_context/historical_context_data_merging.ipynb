```python
health_df.head()
```

Let's select a few that are relevant

```python
GHO_list = ['Cholera case fatality rate', 'Raised blood pressure (SBP>=140 OR DBP>=90) (crude estimate)', 'Total number of notified TB cases','Number of diarrhoea deaths from inadequate water',  'Total fertility rate (per woman)', 'Number of cases of visceral leishmaniasis reported','Number of deaths among children 5 to 14 years of age','Gross national income per capita (Atlas method)']
```

```python
health_df_sub = health_df[['YEAR (DISPLAY)','GHO (DISPLAY)','Numeric']]
```

```python
health_df_sub = health_df_sub.loc[health_df_sub['GHO (DISPLAY)'].isin(GHO_list)]
```

```python
health_df_sub['GHO (DISPLAY)'].unique()
```

```python
health_df_sub['YEAR (DISPLAY)'].unique()
```

```python
health_pivot = pd.pivot_table(health_df_sub, values='Numeric', index=['YEAR (DISPLAY)'], columns=['GHO (DISPLAY)'], aggfunc='sum')
```

```python
health_pivot.head()
```

```python
#health_pivot.columns = health_pivot.columns.droplevel(0)
#health_pivot.columns.name = None
health_final = health_pivot.reset_index()
```

```python
health_final.head()
```

```python
health_final = health_final.rename(columns={"YEAR (DISPLAY)":"Year"})
```

4.) Merging the dataframes together

```python
final_df = comp_conflict_df.merge(eco_df_sub, left_on='year', right_on='Year')
```

```python
final_df = final_df.merge(health_final, left_on='year', right_on='Year')
```
